```python
sns.set(font_scale=1.75)

fig, ax = plt.subplots(figsize=(18,18))

chart = sns.countplot(data=top50, x='Genre', ax=ax)

chart.set_xticklabels(chart.get_xticklabels(), rotation=90)

plt.plot()
```
